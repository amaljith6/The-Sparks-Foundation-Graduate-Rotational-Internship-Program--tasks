```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading data
data = pd.read_csv('SampleSuperstore.csv')
data.head(5)
```

```python
#checking null values
data.isnull().sum()
```

```python
data.describe()
```

```python
#checking correlation
plt.figure(1, figsize = (17, 8))
cor = sns.heatmap(data.corr(), annot = True)
```

```python
data['Country'].unique()
```

There is only one country - united states

###  Top cities in sales 

```python

top_cities= data['City'].value_counts().nlargest(10)
top_cities
```

### Visualising  Top cities in sales

```python
top_cities.plot(kind='bar', color='red')
plt.ylabel('Count')
plt.xlabel('cities')
plt.show()
```

###  Analyzing different catagories

```python
category=data['Category'].unique()
print(category)
```

```python
data['Category'].value_counts()
```

```python
sns.countplot(x='Category',data=data)
```

### Top states

```python

top_states= data['State'].value_counts().nlargest(10)
top_states
```

```python
plt.figure(figsize=(7,5))
top_states.plot(kind='barh', color='blue')
plt.xlabel('Count')
plt.ylabel('states')
plt.show()
```

### Distribution of customer Segment

```python
plt.figure(figsize=(7,5))
sns.countplot("Segment",hue='Segment' ,data = data)          
```

### Sub-Category Wise Sales

```python
plt.figure(figsize=(14,8))
data['Sub-Category'].value_counts().plot.bar(rot=0)
plt.title('Sub-Category Wise Sales')
plt.ylabel('Sales')
plt.xlabel('Sub Categories')
plt.show()
```

Binders and Paper has wide variety

### visualising the count of each subcategory

```python
plt.figure(figsize=(10,10))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

### Visualising Ship Mode

```python
plt.figure(figsize=(7,7))
sns.set_style("whitegrid")
sns.countplot(x='Ship Mode',data=data,palette="Set2")
```

```python
plt.figure(figsize=(7,7))
sns.set_style("darkgrid")
sns.countplot(x="Region",data=data,palette="Set1")
```

## Analyzing the profit of sales 

```python
plt.figure(figsize=(14,12))
statewise=data.groupby(['State'])['Profit'].sum().nlargest(50)
statewise.plot.barh()
```

California state have highest profit and Texas have lowest profit

# Total Profit and sales per category

```python
plt.figure(figsize=(15,10))
s=data.groupby('Sub-Category')['Profit','Sales'].sum()
s.plot.bar( rot=0,title="Total Profit and sales per category",figsize=(15,10));
plt.show(block=True)
```
